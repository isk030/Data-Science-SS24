```python
cleaned_data = airbnb_data.copy()

# Fehlende Werte in 'license' durch 'no' ersetzen
cleaned_data['license'] = cleaned_data['license'].fillna('no',)

# Fehlende Werte in 'house_rules' durch 'no' ersetzen
cleaned_data['house_rules']  = cleaned_data['house_rules'].fillna('no')

# Überprüfen der Änderungen
missing_values_step_1 = cleaned_data[['license', 'house_rules']].isnull().sum()
missing_values_step_1
```

```python
# Fehlende Werte in 'last review' durch 'missing' ersetzen
cleaned_data['last review'] = cleaned_data['last review'].fillna('missing')

# Überprüfen der Änderungen
missing_values_step_2 = cleaned_data['last review'].isnull().sum()
missing_values_step_2
```

```python
# Funktion zum Berechnen des Medians in der gleichen 'neighbourhood group'
def fill_reviews_per_month(row):
    if pd.isnull(row['reviews per month']):
        if row['number of reviews'] > 0:
            median_value = cleaned_data[cleaned_data['neighbourhood group'] == row['neighbourhood group']]['reviews per month'].median()
            if pd.isna(median_value):
                median_value = cleaned_data[cleaned_data['neighbourhood'] == row['neighbourhood']]['reviews per month'].median()
            return median_value
        else:
            return 0
    return row['reviews per month']

# Fehlende Werte in 'reviews per month' ersetzen
cleaned_data['reviews per month'] = cleaned_data.apply(fill_reviews_per_month, axis=1)

# Überprüfen der Änderungen
missing_values_step_3 = cleaned_data['reviews per month'].isnull().sum()
cleaned_data[cleaned_data['reviews per month'].isnull()]
```

```python
# Entfernen von Einträgen mit fehlenden Werten in 'lat' oder 'long'
cleaned_data = cleaned_data.dropna(subset=['lat', 'long'])

# Überprüfen der Änderungen
missing_values_after_dropping = cleaned_data[['lat', 'long']].isnull().sum()
missing_values_after_dropping
```

```python
# Fehlende Werte in 'host name' durch 'anonymous' ersetzen
cleaned_data['host name'] = cleaned_data['host name'].fillna('anonymous')

# Überprüfen der Änderungen
missing_values_step_4 = cleaned_data['host name'].isnull().sum()
missing_values_step_4
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import BallTree

# Konvertiere Längen- und Breitengrade in Radianten
cleaned_data['lat_rad'] = np.radians(cleaned_data['lat'])
cleaned_data['long_rad'] = np.radians(cleaned_data['long'])

# BallTree für effiziente Nachbarschaftssuche erstellen
coords = cleaned_data[['lat_rad', 'long_rad']].values
tree = BallTree(coords, metric='haversine')

# Funktion zum Finden der häufigsten 'neighbourhood group' innerhalb eines Radius von 5 km
def fill_neighbourhood_group(row):
    if pd.isnull(row['neighbourhood group']):
        lat, long = row['lat'], row['long']
        coord_rad = np.radians([lat, long])
        # Suche Nachbarn innerhalb von 5 km (0.005 ist der ungefähre Radius in Radianten)
        indices = tree.query_radius([coord_rad], r=5 / 6371.0)[0]
        nearby = cleaned_data.iloc[indices]
        if not nearby.empty:
            return nearby['neighbourhood group'].mode()[0]
    return row['neighbourhood group']

# Fehlende Werte in 'neighbourhood group' ersetzen
cleaned_data['neighbourhood group'] = cleaned_data.apply(fill_neighbourhood_group, axis=1)

# Überprüfen der Änderungen
missing_values_step_9 = cleaned_data['neighbourhood group'].isnull().sum()
missing_values_step_9
```

```python
# Funktion zum Berechnen des Medians in der gleichen 'neighbourhood group'
def fill_minimum_nights(row):
    if pd.isnull(row['minimum nights']):
        median_value = cleaned_data[cleaned_data['neighbourhood group'] == row['neighbourhood group']]['minimum nights'].median()
        return median_value
    return row['minimum nights']

# Fehlende Werte in 'minimum nights' ersetzen
cleaned_data['minimum nights'] = cleaned_data.apply(fill_minimum_nights, axis=1)

# Überprüfen der Änderungen
missing_values_step_5 = cleaned_data['minimum nights'].isnull().sum()
missing_values_step_5
```

```python
# Fehlende Werte in 'calculated host listings count' durch 1 ersetzen, wenn 'host_identity_verified' nicht NaN ist, ansonsten Eintrag entfernen
cleaned_data['calculated host listings count'] = cleaned_data.apply(
    lambda row: 1 if pd.isnull(row['calculated host listings count']) and pd.notnull(row['host_identity_verified']) else row['calculated host listings count'], axis=1
)
cleaned_data = cleaned_data[cleaned_data['calculated host listings count'].notnull()]

# Überprüfen der Änderungen
missing_values_step_6 = cleaned_data['calculated host listings count'].isnull().sum()
missing_values_step_6
```

```python
# Funktion zum Berechnen des Medians in der gleichen 'neighbourhood group'
def fill_availability_365(row):
    if pd.isnull(row['availability 365']):
        median_value = cleaned_data[cleaned_data['neighbourhood group'] == row['neighbourhood group']]['availability 365'].median()
        return median_value
    return row['availability 365']

# Fehlende Werte in 'availability 365' ersetzen
cleaned_data['availability 365'] = cleaned_data.apply(fill_availability_365, axis=1)

# Überprüfen der Änderungen
missing_values_step_7 = cleaned_data['availability 365'].isnull().sum()
missing_values_step_7
```

```python
# Fehlende Werte in 'host_identity_verified' durch 'verified' ersetzen, wenn 'number of reviews' > 9, ansonsten 'unconfirmed'
cleaned_data['host_identity_verified'] = cleaned_data.apply(
    lambda row: 'verified' if pd.isnull(row['host_identity_verified']) and row['number of reviews'] > 9 else 'unconfirmed' if pd.isnull(row['host_identity_verified']) else row['host_identity_verified'],
    axis=1
)

# Überprüfen der Änderungen
missing_values_step_8 = cleaned_data['host_identity_verified'].isnull().sum()
missing_values_step_8
```

```python

# Überprüfen auf unplausible Werte in den angegebenen Spalten

# Construction year: Negative Werte oder Werte größer als 2024
invalid_construction_year_before = cleaned_data[(cleaned_data['Construction year'] < 0) | (cleaned_data['Construction year'] > 2024)].shape[0]

# Price: keine negativen Werte
# Zuerst sicherstellen, dass die Spalte price numerisch ist
cleaned_data['price'] = pd.to_numeric(cleaned_data['price'], errors='coerce')
invalid_price_before = cleaned_data[(cleaned_data['price'] < 0)].shape[0]

# Entfernen der unplausiblen Einträge
cleaned_data = cleaned_data[(cleaned_data['Construction year'] >= 0) & (cleaned_data['Construction year'] <= 2024)]
cleaned_data = cleaned_data[cleaned_data['price'] >= 0]

# Überprüfen, ob die unplausiblen Werte entfernt wurden
invalid_construction_year_after = cleaned_data[(cleaned_data['Construction year'] < 0) | (cleaned_data['Construction year'] > 2024)].shape[0]
invalid_price_after = cleaned_data[(cleaned_data['price'] < 0)].shape[0]

# Ergebnisse anzeigen
invalid_entries_before = {
    'Invalid Construction Year Before': invalid_construction_year_before,
    'Invalid Price Before': invalid_price_before,
}

invalid_entries_after = {
    'Invalid Construction Year After': invalid_construction_year_after,
    'Invalid Price After': invalid_price_after,
}

print("Anzahl unplausibler Einträge vor der Bereinigung:")
print(invalid_entries_before)

print("Anzahl unplausibler Einträge nach der Bereinigung:")
print(invalid_entries_after)
```

```python
# Überprüfen auf negative Werte in numerischen Spalten, aber ohne die Spalte 'long'
numerical_columns = cleaned_data.select_dtypes(include=['float64', 'int64']).columns
numerical_columns = numerical_columns.drop('long')  # Spalte 'long' ausschließen
numerical_columns = numerical_columns.drop('long_rad')  # Spalte 'long' ausschließen

negative_values = cleaned_data[(cleaned_data[numerical_columns] < 0).any(axis=1)]

print("Einträge mit negativen Werten in numerischen Spalten (ohne 'long'):")
print(negative_values)

# Überprüfen auf Tippfehler in string-basierten Spalten
string_columns = cleaned_data.select_dtypes(include=['object']).columns

# Funktion zur Überprüfung auf mögliche Tippfehler
def find_possible_typos(column):
    value_counts = column.value_counts()
    possible_typos = value_counts[value_counts <= 5]  # Annahme: Werte mit sehr wenigen Vorkommen könnten Tippfehler sein
    return possible_typos

# Überprüfung der string-basierten Spalten auf mögliche Tippfehler
for col in string_columns:
    typos = find_possible_typos(cleaned_data[col])
    if not typos.empty:
        print(f"Mögliche Tippfehler in der Spalte '{col}':")
        print(typos)

# Beispielhafte Anzeige der Ergebnisse
print(cleaned_data.head())
```

```python
# Mapping von Tippfehlern zu den richtigen Werten
correction_mapping = {
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan'
}

# Tippfehler in 'neighbourhood group' korrigieren
cleaned_data['neighbourhood group'] = cleaned_data['neighbourhood group'].replace(correction_mapping)

# Tippfehler in 'neighbourhood' korrigieren
cleaned_data['neighbourhood'] = cleaned_data['neighbourhood'].replace(correction_mapping)

# Überprüfen der Änderungen
print("Einzigartige Werte in 'neighbourhood group':")
print(cleaned_data['neighbourhood group'].unique())

print("\nEinzigartige Werte in 'neighbourhood':")
print(cleaned_data['neighbourhood'].unique())

# Speichern der bereinigten Daten
cleaned_data.to_csv('Airbnb_Open_Data_Cleaned.csv', index=False)
```
